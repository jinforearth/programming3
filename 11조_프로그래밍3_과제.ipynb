<h3><font color=red>2017년도</font>와 <font color=red>2018년도</font> 데이터의 경우에는 <font color=red>컬럼이 <b>한글</b>로 구성</font>되어 있다는 것을 알 수 있습니다.</h>
<h3>따라서, 한글로 된 <font color=red>컬럼명을 <b>영어</b></font>로 고치는 작업을 실행합니다.</h>

<h3>-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</h>

<h3><font color=red>방법 1.</font> &nbsp&nbsp 2017, 2018년도의 데이터 컬럼을 각각 수정하기.</h>

<h3>2017년도 데이터 컬럼 수정</h>

```python
air_hour_2017.columns = ['data_dt', 'loc_code', 'item_code',
                'data_value', 'data_state', 'data_nover', 'data_rover']
air_hour_2017
```

<h3>2018년도 데이터 컬럼 수정</h>

```python
air_hour_2018.columns = ['data_dt', 'loc_code', 'item_code',
                'data_value', 'data_state', 'data_nover', 'data_rover']
air_hour_2018
```

<h3>수정한 데이터를 df_1로 합치기</h>

```python
df_1 = pd.concat([air_hour_2017,air_hour_2018], keys=[2017,2018])
print(df_1)
```

<h3><font color=red>방법 2.</font> &nbsp&nbsp 2017, 2018년도의 데이터를 df_1로 합친 후 컬럼명을 통합하여 수정하기.</h>

<h3>2017, 2018년도 데이터를 df_1로 합치기</h>

```python
df_1 = pd.concat([air_hour_2017,air_hour_2018], keys=[2017,2018])
print(df_1)
```

<h3>df_1에서 컬럼명을 한 번에 수정하기</h>

```python
df_1.columns = ['data_dt', 'loc_code', 'item_code',
                'data_value', 'data_state', 'data_nover', 'data_rover']
df_1
```

<h3>-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</h>

<h3>2017년도와 2018년도 데이터를 df_1로 합쳤으니, 나머지 2019년도~2021년도의 데이터도 df_2로 합쳐줍시다!</h>

```python
df_2 = pd.concat([air_hour_2019,air_hour_2020,air_hour_2021], keys=[2019,2020,2021])
print(df_2)
```

<h3><font color=red><b>!!!</b></font> df_2의 경우에는 컬럼명이 영어로 되어있는, 2017년도와 2018년도의 데이터에는 <font color=red><b>없었던!</b></font></h>
<h3><font color=blue><b>새로운 컬럼</b></font>이 두 개 추가되어 있습니다. <font color=red><b>!!!</b></font></h>
<h3><font color=red>따라서, 두 데이터를 합치기 위해 <font color=black>앞선 2번째 방법으로</font> <b>컬럼명을 통일</b>해 주었습니다.</font></h>

```python
df_1.columns = ['data_dt', 'loc_code', 'item_code',
                'data_value', 'data_state', 'data_nover', 'data_rover']
df_1
```

<h3>-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</h>

<h3>컬럼명을 영어로 수정한 df_1과 df_2를 합치니 data_rover와 regist_dt 에 NaN값이 생기는 상태로 합쳐진 것을 알 수 있습니다.</h>

```python
air_df = pd.concat([df_1, df_2])
air_df
```

<h3>loc_code(측정소 코드), data_value(평균값)을 제외한 나머지는 -128~127 사이에 있으므로 int8로 바꿔주며,</h>

<h3>data_value(평균값)은 소숫점 첫째자리까지 쓰고 있으므로 float16으로 바꿔준다.</h>

```python
air_2021.info()
```

<h3>측정일시에 따른 대기오염 평균값의 평균 구하기</h>

```python
(
    air_2021
    .groupby('data_dt')
    .agg({'data_value':'mean'})
)
```

<h3>측정일시 / 측정항목 코드별 대기오염 평균값 구하기</h>

```python
(
    air_2021
    .groupby(['data_dt', 'item_code'])
    ['data_value']
    .agg('mean')
)
```

<h3>측정일시 / 측정항목 코드별 대기오염 평균값의 평균, 최소, 최대를 구하고 행의 다중 인덱스 제거하기</h>

```python
(
    air_2021
    .groupby(['data_dt', 'item_code'])
    .agg(평균=pd.NamedAgg(column='data_value', aggfunc='mean'),
         최소=pd.NamedAgg(column='data_value', aggfunc='min'),
         최대=pd.NamedAgg(column='data_value', aggfunc='max')
        )
    .reset_index()
)
```
